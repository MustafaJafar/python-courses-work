```python
import requests
from bs4 import BeautifulSoup

#get web page
req = requests.get("https://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
#load content
cont = req.content

soup = BeautifulSoup(cont , "html.parser")
#print(soup.prettify())

#understad the structure of the web page
#use the inspect tool in you web browser
#each search reasult is a div with class property row 
#price in h4 
all = soup.find_all("div" , {"class" : "propertyRow"})
#print(all[0])

prices = all[0].find_all("h4" , {"class" : "propPrice"})
print(prices[0].text)
```

```python
print ( "Prices    ,  Addresses                                     , beds      , area , bath")
print ( "------------------------------------------------------------------------------------")
for item in all : 
    prices = item.find_all("h4" , {"class" : "propPrice"})
    addresses = item.find_all("span" , {"class" : "propAddressCollapse"})
    try :
        beds = item.find_all("span" , {"class" : "infoBed"})[0].text
    except : 
        beds = None
    try :
        area = None
    except : 
        area = item.find_all("span" , {"class" : "infoSqft"})[0].text
    try : 
        bath = item.find_all("span" , {"class" : "infoValueFullBath"})[0].text
    except :
        bath = None
    print(prices[0].text.replace("\n" ,"") , " , " , addresses[0].text, "," , addresses[1].text ," ,", beds , 
            " ," , area , " ," , bath )
```

```python
for item in all : 
    for column_group in item.find_all("div" , {"class" : "columnGroup"}) : 
        for feature_group , feature_name in zip (column_group.find_all("span" , {"class" : "featureGroup"}) ,column_group.find_all("span" , {"class" : "featureName"}) ) :
            print(feature_group.text , feature_name.text)
    print("----------")
```

```python
#put data in a data frame
#1.save each elment as dictionay
#2.list of dictionaries 
l = []
for item in all : 
    d = {}
    
    d["Address"] = item.find_all("span" , {"class" : "propAddressCollapse"})[0].text
    d["Locality"] = item.find_all("span" , {"class" : "propAddressCollapse"})[1].text
    d["Price"] = item.find_all("h4" , {"class" : "propPrice"})[0].text.replace("\n" ,"")
    
    try :
        d["Beds"] = item.find_all("span" , {"class" : "infoBed"})[0].text
    except : 
        d["Beds"]= None
    try :
        d["Area"] = None
    except : 
        d["Area"] = item.find_all("span" , {"class" : "infoSqft"})[0].text
    try : 
        d["Full Baths"] = item.find_all("span" , {"class" : "infoValueFullBath"})[0].text
    except :
        d["Full Baths"] = None
    try : 
        d["Half Baths"] = item.find_all("span" , {"class" : "infoValueHalfBath"})[0].text
    except :
        d["Half Baths"] = None
   
    for column_group in item.find_all("div" , {"class" : "columnGroup"}) : 
        for feature_group , feature_name in zip (column_group.find_all("span" , {"class" : "featureGroup"}) ,column_group.find_all("span" , {"class" : "featureName"}) ) :
            if ("Lot Size" in feature_group.text) : 
                d["Lot Size"] = feature_name.text
    l.append(d)
```

```python
import pandas 
df = pandas.DataFrame(l)
```
